import the libraries

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras  

print (tf.__version__)
```

```python
#define a simple model with one neuron  which is depicted by the "Dense"  and one layer depicted by sequential having one entry, there is only one input value.

model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
# Compile the model with two functions: Loss and Optimizer. Loss will check if the predicted value is closer to actual value. then uses optimizer function to guess a new answer.

model.compile(optimizer='sgd', loss='mean_squared_error')
```

when i used Adam as the optimizer , i got the loss as loss: 2.6133 and predicted value as 10, which shows deviation from actual value. need to find the reason.

then i ran using RMS Prop and preidcted value was 15 better than Adam but not as accurate as SGD

```python
#what if i use adam as optimizer
model.compile(optimizer='RMSprop', loss='mean_squared_error')
```

```python
#provide the input
xs = np.array([-1.0,0.0,1.0,2.0,3.0,4.0],dtype = float)
ys = np.array([-3.0,-1.0,1.0,3.0,5.0,7.0],dtype = float)
```

```python
#train the model for 500 epochs meaning model will run the 500 times with the same data to reduce the loss and find the pattern

model.fit(xs,ys,epochs=500)
```

```python
#once the model is run, lets make the prediction
print(model.predict([10]))
```
